```python
# BAR Chart

students = [1,2,3,4,5]
colors = ['red','green','blue','black','yellow']


plt.bar(colors,students,color='black')
plt.xlabel('colors')
plt.ylabel('no_of_students')
plt.show()
```

```python
#subplots

df = pd.read_csv('batter.csv')
```

```python
df
```

```python
fig , ax = plt.subplots()

ax.scatter(df['avg'],df['strike_rate'])
ax.set_xlabel('avg')
ax.set_ylabel('strike_rate')
ax.set_title('avg VS SR')
fig.show()
```

```python
fig , ax = plt.subplots(nrows = 2 ,ncols=1,figsize=(10,10))

ax[0].scatter(df['avg'],df['strike_rate'])
ax[1].scatter(df['avg'],df['runs'])

ax[0].set_xlabel('avg')
ax[0].set_ylabel('strike_rate')

ax[1].set_xlabel('avg')
ax[1].set_ylabel('Runs')

ax[0].set_title('avg VS SR')
ax[1].set_title('avg VS Runs')
fig.show()
```

```python
#3D plots
#3D scatter plot
df = pd.read_csv('batter.csv')

ax = plt.subplot(projection = '3d')
ax.scatter3D(df['runs'],df['avg'],df['strike_rate'])
ax.set_title('runs VS avg VS strike_rate')
ax.set_xlabel('runs ')
ax.set_ylabel('avg ')
ax.set_zlabel('strike_rate')

```

```python
#3D line plot

x = [0,1,5]
y = [10,5,6]
z = [5,10,15]
ax = plt.subplot(projection='3d')
ax.scatter(x,y,z)
ax.plot3D(x,y,z,color='red')
```

```python
#3D surface plot

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx , yy = np.meshgrid(x,y)

z = xx**2 + yy**2


fig = plt.figure(figsize=(10,10))
ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z)

```
