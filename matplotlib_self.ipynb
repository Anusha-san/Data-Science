```python
df=pd.read_csv('sharma-kohli.csv')
```

```python
df
```

```python

plt.plot(df['index'],df['RG Sharma'])
plt.show()
```

```python
plt.plot(df["index"],df["V Kohli"],color="#ff9900",linestyle="dashed",marker="D",markersize=10)
plt.plot(df['index'],df['RG Sharma'],color="#660066",linestyle="dotted",marker="o",markersize=10)
plt.xlabel("YEARS")
plt.ylabel("TOTAL RUNS")
plt.title("Years VS Runs Scored")
plt.show()
```

```python
plt.plot(df["index"],df["V Kohli"],color="red")
plt.xlabel("YEARS")
plt.ylabel("TOTAL RUNS")
plt.title("Years VS Runs Scored")
plt.show()
```

```python
plt.plot(df["index"],df["V Kohli"],color="#ff9900",linestyle="dashed",marker="D",markersize=10,label="V Kohli")
plt.plot(df['index'],df['RG Sharma'],color="#660066",linestyle="dotted",marker="o",markersize=10,label="RG Sharma")

plt.xlabel("YEARS")
plt.ylabel("TOTAL RUNS")
plt.title("Years VS Runs Scored")
plt.legend(loc="best")
plt.show()

```
