```python
df = pd.read_csv('/content/sharma-kohli.csv')
```

```python
df
```

```python
plt.style.use('default')

plt.plot(df['index'],df['RG Sharma'])
```

```python
plt.plot(df['index'],df['V Kohli'])

plt.xlabel('YEARS')
plt.ylabel('TOTAL RUNS')
plt.title('Year VS Runs Scored')
```

```python
plt.plot(df['index'],df['V Kohli'],color='red')
plt.plot(df['index'],df['RG Sharma'],color='#ff66ff')

plt.xlabel('YEARS')
plt.ylabel('TOTAL RUNS')
plt.title('Year VS Runs Scored')
```

```python
plt.plot(df['index'],df['V Kohli'],color='red',linestyle= 'dashed')
plt.plot(df['index'],df['RG Sharma'],color='#ff66ff',linestyle= 'dotted')

plt.xlabel('YEARS')
plt.ylabel('TOTAL RUNS')
plt.title('Year VS Runs Scored')
```

```python
#color,marker,linestyle

plt.plot(df['index'],df['V Kohli'],color='red',linestyle= 'dashed',marker='D',markersize=10)
plt.plot(df['index'],df['RG Sharma'],color='green',linestyle= 'dotted',marker='o',markersize=15)

plt.xlabel('YEARS')
plt.ylabel('TOTAL RUNS')
plt.title('Year VS Runs Scored')
```

```python
#legends->location
plt.plot(df['index'],df['V Kohli'],color='red',linestyle= 'dashed',marker='D',markersize=10,label='V Kohli')
plt.plot(df['index'],df['RG Sharma'],color='green',linestyle= 'dotted',marker='o',markersize=15,label='R Sharma')

plt.legend()
plt.grid()

plt.xlabel('YEARS')
plt.ylabel('TOTAL RUNS')
plt.title('Year VS Runs Scored')
```
